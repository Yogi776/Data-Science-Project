```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import transformers
from keras.preprocessing import sequence
import tqdm
```

```python
df  =  pd.read_csv('GrammarandProductReviews.csv')
```

```python
df
```

```python
df.columns
```

```python
df1 = df[['reviews.text','reviews.rating']]
```

```python
df1
```

```python
df1.dropna(inplace=True)
```

```python
df1.isnull().sum()
```

```python
from transformers import BertTokenizer
bert_tokenizer = transformers.RobertaTokenizer.from_pretrained('roberta-base')
```

```python
def func_tokenizer(tokenizer_name,docs):
    features = []
    for doc in tqdm.tqdm(docs,desc= "Converting documents to features"):
        tokens = tokenizer_name.tokenize(doc)
        ids = tokenizer_name.convert_tokens_to_ids(tokens)
        features.append(ids)
    return features
```

```python
bert_features = func_tokenizer(bert_tokenizer,df1['reviews.text'])
```

```python
bert_trg = sequence.pad_sequences(bert_features,maxlen=500)
```

```python
bert_trg
```

```python
y = df1['reviews.rating'].values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(bert_trg,y,test_size=0.2,random_state= 3)
```

```python
model1 = LogisticRegression(solver='saga',penalty = 'l1')
```

```python
model1.fit(x_train,y_train)
```

```python
model1.score(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```
