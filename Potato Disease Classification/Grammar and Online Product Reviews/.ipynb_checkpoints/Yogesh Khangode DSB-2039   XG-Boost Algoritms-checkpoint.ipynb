```python
ps = PorterStemmer()
```

```python
def tranform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text) 
    
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)
            
            
    text = y[:]
    y.clear()
    
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
            
    text = y[:]
    y.clear()
    
    for i in text:
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
df1['transformed_text'] = df1['Reviews'].apply(tranform_text)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
bnb  = BernoulliNB()
svc  = SVC(kernel='sigmoid',gamma=1.0)
knc  = KNeighborsClassifier()
mnb  = MultinomialNB()
dtc  = DecisionTreeClassifier(max_depth=5)
lrc  = LogisticRegression(solver='liblinear',penalty = 'l1')
rfc  = RandomForestClassifier(n_estimators=100,random_state=2)
abc  = AdaBoostClassifier(n_estimators=100,random_state=2)
etc  = ExtraTreesClassifier(n_estimators=100,random_state=2)
gbdt = GradientBoostingClassifier(n_estimators=100,random_state=2)
xgb  = XGBClassifier(n_estimators=100,random_state=2)
```

```python
clfs = {
    'bnb':bnb,
    'SVC':svc,
    'KNC':knc,
    'NB':mnb,
    'DT':dtc,
    'LR':lrc,
    'AdaBoost':abc,
    'Bgc':abc,
    'Etc':etc,
    'GBDT':gbdt,
    'xgb':xgb
}
```

```python
def train_classifier(clf,x_train,y_train,x_test,y_test):
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred)
    training_acc = clf.score(x_train,y_train)
    return accuracy,training_acc
```

```python
accuracy_scores = []
training = []
for name,clf in clfs.items():
    current_accuracy,model_traing= train_classifier(clf,x_train,y_train,x_test,y_test)
    print("Model Name:-              ",name)
    print("Model Traing Accuracy :-  ",model_traing)
    print("Model  Test  Accuracy :-  ",current_accuracy)
    accuracy_scores.append(current_accuracy) 
    training.append(model_traing)
```

```python
performance_df1 = pd.DataFrame({"Algorithms":clfs.keys(), "Training Accuracy":training,'Testing Accuracy':accuracy_scores})
```

## MultinomialNB

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import RegexpTokenizer

#tokenizer to remove unwanted elements from out data like symbols and numbers
token = RegexpTokenizer(r'[a-zA-Z0-9]+')

tf=TfidfVectorizer(lowercase=True,
                     stop_words='english',
                     ngram_range = (1,2),
                     tokenizer = token.tokenize)

text_counts= tf.fit_transform(df1['Reviews'])

X_train, X_test, y_train, y_test = train_test_split(text_counts, df1['Rating'], test_size=0.3, random_state=1)

clf = MultinomialNB()

clf.fit(X_train, y_train)
print("Training Accuracy: ",clf.score(X_train,y_train))
print("Testing Accuracy : ",clf.score(X_test,y_test))
```
